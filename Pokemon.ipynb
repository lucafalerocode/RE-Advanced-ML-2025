```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/Advanced ML/Data Sets/pokemons.csv")
```

## 1. What are the most common Pokémon types?

```python
# Count the occurrences of each type
type_counts = df['type1'].value_counts()
print(type_counts)
```

## 2. How do different stats compare across generations?

```python
# Group by generation and get mean values for stats
gen_stats = df.groupby('generation')[['hp', 'atk', 'def', 'spatk', 'spdef', 'speed']].mean()
print(gen_stats)
```

## 3. Is there a correlation between a Pokémon's weight and its total stats?

```python
# Calculate correlation between weight and total stats
correlation = df['weight'].corr(df['total'])
print(f"Correlation between weight and total stats: {correlation}")
```
